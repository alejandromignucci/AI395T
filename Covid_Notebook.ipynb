```python
import pandas as pd
```

```python
# Read final_df.csv into a DataFrame
df = pd.read_csv('final_df.csv')

df.head()
```

```python
# Create a Mean Arterial Pressure (MAP) column
df['MAP'] = (df['Systolic Blood Pressure'] + 2*df['Diastolic Blood Pressure'])/3

df.head()
```

```python
# Drop columns that are not needed
df = df.drop(columns=['Systolic Blood Pressure', 'Diastolic Blood Pressure'])

df.head()
```

```python
# Export the DataFrame to a new CSV file
df.to_csv('map_df.csv', index=False)
```
